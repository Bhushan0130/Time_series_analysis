## Time Resampling

```python
df_tsla.resample(rule ='A' ).min()
```

```python
df_tsla.resample(rule ='A' ).max()
```

```python
type(df_tsla.resample(rule ='A' ).max())
```

```python
df_tsla.resample(rule ='A' ).max()['High'].plot(figsize = (12,4))
```

```python
# quaterly start frequency 
df_tsla.resample(rule ='QS' ).min()
```

```python
df_tsla.resample(rule ='QS' ).max()['High'].plot(figsize = (12,4))
```

```python
# Business end frequency
df_tsla.resample(rule ='BA' ).max()
```

```python
df_tsla.resample(rule ='BA' ).max()['High'].plot(figsize =(12,4))
```

```python
## plotting
df_tsla['High'].resample(rule ='BA' ).mean().plot(kind = 'bar', figsize =(12,4))
```

```python
## Rolling
df_tsla['High'].rolling(11).mean().head(20)
```

```python
df_tsla['Open: 30 Days Rolling'] = df_tsla['Open'].rolling(11).mean()
```

```python
df_tsla.head()
```

```python
df_tsla[['Open', 'Open: 30 Days Rolling' ]].plot(figsize = (12,4))
```
